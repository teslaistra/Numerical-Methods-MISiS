Импорт нужных внешних модулей

```python
import pandas as pd
import math
import plotly.express as px
import numpy as np
import pylab as pl
import plotly
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
import docx
```

Объявление всех нужных констант для работы. Отрезка на котором ищем корень и точности.

```python
START_X = -math.pi/2
FINISH_X = -0.1
EPS = [0.01, 0.001, 0.0001]
```

Реализация самой функции для которой ищется корень

```python
def func(x):
    return (4.0 * math.cos(x) - x - (1.0/float(math.pow(x,2))))
```

Реализация метода ложных позиций

```python
def false_position(func, df, x0, x1, eps=0.0001):
    delx = x1-x0

    f0 = func(x0)
    f1 = func(x1)

    while delx > eps:
        xm = x1*f0/(f0-f1) + x0*f1/(f1-f0)
        fm = func(xm)
        
        if f0*fm > 0:
            delx = xm-x0
            x0 = xm
            f0 = fm
        else:
            delx = x1-xm
            x1 = xm
            f1 = fm
        df.loc[len(df)] = [x0, f0, x1, f1, delx]
    return df
```

Цикл в котором мы вызываем алгоритм для разной степени точности и выводим результаты.

```python
for eps in EPS:
    df = pd.DataFrame(columns=['Xleft', 'f(Xleft)',
                               'Xright', 'f(Xright)', 'delta'])
    df = false_position(func, df, START_X, FINISH_X, eps)
    print(f"Finding root at [{START_X}; {FINISH_X}] with eps = {eps}")
    print(df)
    print(
        f"Root is {(df[-1:]['Xleft'].values[0])}")
    print("_______________________________________________________________")
```

```python
# open an existing document
doc = docx.Document('test.docx')

# add a table to the end and create a reference variable
# extra row is so we can add the header row
t = doc.add_table(df.shape[0]+1, df.shape[1])

for j in range(df.shape[-1]):
    t.cell(0, j).text = df.columns[j]

# add the rest of the data frame
for i in range(df.shape[0]):
    for j in range(df.shape[-1]):
        t.cell(i+1, j).text = str(df.values[i, j])

# save the doc
doc.save('test.docx')
```
