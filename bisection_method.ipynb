```python
import pandas as pd
import math
```

```python
def func(x):
    return (4.0 * math.cos(x) - x - (1.0/float(math.pow(x,2))))
```

```python
def bisection_method(func, df, x0=0.2, x1=1.2, eps=0.0001):
    x0 = 0.2
    x1 = 1.2
    eps = 0.0001
    delx = x1-x0
    f0 = func(x0)
    f1 = func(x1)
    while delx > eps:
        xm = 0.5*(x0+x1)
        fm = func(xm)
        if f0*fm > 0:
            x0 = xm
            f0 = fm
        else:
            x1 = xm
            f1 = fm
        delx = math.fabs(x1-x0)
        df.loc[len(df)] = [x0, f0, x1, f1, delx]
    return df
```

```python
df = pd.DataFrame(columns=['Xleft','f(Xleft)','Xright','f(Xright)','delta'])
```

```python
df = bisection_method(func,df,-math.pi,math.pi)
```
